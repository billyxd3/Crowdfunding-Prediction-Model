```python
df = pd.read_csv("../data/cleaned_data/kickstarter.csv")
df.head()
```

```python
# Calculate the percentage of each campaign status
status_counts = df['state'].value_counts(normalize=True) * 100

# Plotting a pie chart
plt.figure(figsize=(8, 8), facecolor='lightgray')
plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', startangle=90, textprops={'color': 'black'})
plt.title('Campaigns Success Ratio', color='green')

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Display the chart
plt.show()
print(df.state.value_counts())
target = df.state.value_counts(normalize=True) 
print(target)
```

<h2> 2. Country </h2>

```python
country_list = df.country.value_counts()
plt.figure(figsize=(16,5))
ax = sns.barplot(x=country_list.index, y=country_list.values, alpha=0.8)
ax.xaxis.label.set_color('black')
ax.yaxis.label.set_color('black')
ax.tick_params(axis='x', colors='black')
ax.tick_params(axis='y', colors='black')
plt.title('Kickstarter Countries', color='black')
plt.ylabel('Number of Campaigns', fontsize=12)
```

```python
df['state'] = df.state.astype(str)
df['success'] = np.where(df.state == "successful", 1, 0)
df['fail'] = np.where(df.state == "failed", 0, 1)
df.success.value_counts()
df.fail.value_counts()
```

```python
country_success = df[df.success == 1].groupby(['country']).size().sort_values(ascending=False)

plt.figure(figsize=(16,5))
ax = sns.barplot(x=country_success.index, y=country_success.values)
ax.xaxis.label.set_color('black')
ax.yaxis.label.set_color('black')
ax.tick_params(axis='x', colors='black')
ax.tick_params(axis='y', colors='black')
plt.xticks(rotation=45)
plt.title('Successful Kickstarter Projects by Country', color='black')
plt.xlabel('Country')
plt.ylabel('Number of Successful Projects')
```

```python
country_fail = df[df.fail == 0].groupby(['country']).size().sort_values(ascending=False)

plt.figure(figsize=(16,5))
ax = sns.barplot(x=country_fail.index, y=country_fail.values)
ax.xaxis.label.set_color('black')
ax.yaxis.label.set_color('black')
ax.tick_params(axis='x', colors='black')
ax.tick_params(axis='y', colors='black')
plt.xticks(rotation=45)
plt.title('Failed Kickstarter Projects by Country', color='black')
plt.xlabel('Country')
plt.ylabel('Number of Failed Projects')
```

```python
country_dict = {'AU': 'OTHER','CA': 'OTHER', 'DE':'OTHER', 'ES': 'OTHER', 'FR': 'OTHER',
 'GB': 'OTHER', 'HK': 'OTHER', 'IT': 'OTHER', 'MX': 'OTHER', 'NL': 'OTHER', 'SE': 'OTHER'}
df = df.replace({"country": country_dict})
```

```python
country_success = df[df.success == 1].groupby(['country']).size()
print(country_success)


plt.figure(figsize=(16,5))
ax = sns.barplot(x=country_success.index, y=country_success.values)
ax.xaxis.label.set_color('black')
ax.yaxis.label.set_color('black')
ax.tick_params(axis='x', colors='black')
ax.tick_params(axis='y', colors='black')
plt.xticks(rotation=45)
plt.title('Successful Kickstarter Projects USA and Other countries', color='black')
plt.xlabel('Country')
plt.ylabel('Number of Successful Projects')
```

```python
# Make dummies
df = pd.get_dummies(columns=['country'], drop_first=True, data=df)
df.columns
```

<h2> 3. Campaign Duration </h2>

```python
df['created_at'] = pd.to_datetime(df.created_at)
df['launched_at'] = pd.to_datetime(df.launched_at)
df['deadline'] = pd.to_datetime(df.deadline)

prep_date = df.created_at.values
start_date = df.launched_at.values
end_date = df.deadline.values

# campaign duration = from launched time to deadline
df['cam_duration'] = pd.to_timedelta(end_date - start_date).days
df['prep_duration'] = pd.to_timedelta(start_date - prep_date).days
```

```python
#sanity check
df[['created_at', 'launched_at', 'deadline', 'state_changed_at', 'cam_duration', 'prep_duration']].head()
```

<h2> 4. Spotlight and Staff pick </h2>

```python
# spotlight and staff_pick
df = pd.get_dummies(columns = ['spotlight','staff_pick'], drop_first=True, data=df)
df.sample(5)
```

<h2> 5. Main Category </h2>

```python
# main category + success
category_success = df[df.success == 1].groupby(['main_category']).size()
print(category_success)
plt.figure(figsize=(16,5))
ax = sns.barplot(x=category_success.index, y=category_success.values)
ax.xaxis.label.set_color('black')
ax.yaxis.label.set_color('black')
ax.tick_params(axis='x', colors='black')
ax.tick_params(axis='y', colors='black')
plt.xticks(rotation=45);
plt.title('Successful Projects by Categories', color='black')
```

```python
# Make category dummies
df = pd.get_dummies(columns=['main_category'], drop_first=True, data=df)
df.columns
```

<h2> 7. Blurb(description of the project) </h2>

```python
df.blurb.head(5)
```

```python
blurb = df.blurb.astype('str')
description_length = []

for b in blurb:
    description_length_item = len(b.split(' '))
    description_length.append(description_length_item)

df['description_length'] = description_length
```

```python
df.rename(columns={'spotlight_1': 'spotlight_True'}, inplace=True)
df.rename(columns={'staff_pick_1': 'staff_pick_True'}, inplace=True)
```
